```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs 
```

```python
data = make_blobs(n_samples=200,
                 n_features=2,
                 centers=4,
                 cluster_std=1.8,
                 random_state=101) # Generate isotropic Gaussian blobs for clustering.
```

```python
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4) # 4 clusters are in the generated data
```

```python
kmeans.fit(data[0]) # features == data[0]
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_ # labels of the clusters
```

```python
data[1] # these are the correct labels of the data
# k-means is unsupervised learning to this would not be available in a real project
```

```python
fig , (ax1,ax2) = plt.subplots(1,2,sharey=True,figsize=(10,6))
ax1.set_title('K Means') # colored by the predicted kmeans.labels_
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='rainbow')
ax2.set_title('Original data') # colored by the factual clusters
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```
